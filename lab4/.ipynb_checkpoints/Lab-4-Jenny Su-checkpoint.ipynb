```python
# Exercise 1
import pandas
import matplotlib.pyplot as plt
from scipy import stats
import numpy

scores = pandas.read_csv('Dataset student_score.csv')
r = stats.linregress(scores.Xi,scores.Yi)
res = numpy.linspace(min(scores.Xi), max(scores.Xi)).reshape(-1,1)
print("The Linear Regression equation is f(x)=" + ('{0:.2f}').format(r[0]) + "x+" + ('{0:.2f}').format(r[1]))
print("\tx is the score and f(x) is the statisics grade")
print("\nIf a student were to get an 80 on the test")
print("\tf(80)=" + ('{0:.2f}').format(r[0]*80+r[1])+"%")
print("they are expected to have ~78.29% for his statistics grade")
print("\nI think it fits because the std_err (average vertical distance) is small: " + ('{0:.2f}').format(r[3]))
print("\tand the r_squared value (Goodness-of-fit) is small: " + ('{0:.2f}').format(r[2]**2)+ ", but the line looks okay and the sample size is small.")
```

```python
# Exercise 2
plt.scatter(scores.Xi, scores.Yi)
plt.title('Test Score vs Statistics Grade')
plt.plot(res, r[1] + r[0]*res, 'r')
plt.ylabel("Statistics Grade")
plt.xlabel("Test Score")
plt.show()
```

```python
# Exercise 3
print("The R-squared value is: " + ('{0:.6f}').format(r[2]**2))
```

```python
# Exercise 5
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score
import seaborn

file = pandas.read_csv('Dataset heart-disease-dataset.csv')
data = datasets.load_diabetes()
dp = pandas.DataFrame(data.data, columns = data.feature_names)
dp['bloodSugar'] = data.target

correlation_matrix = dp.corr().round(2)
seaborn.set(rc={'figure.figsize':(11.7,8.27)})
seaborn.heatmap(data=correlation_matrix, annot=True)
print("Anazlysing the heatmap, bmi seems to have a strongest coorelation to blood sugar with 0.59")
print("So we will be analyzing the regression between bmi and blood sugar")
```

```python
x = dp.bmi
y = dp.bloodSugar

x = numpy.array(x).reshape(-1,1)
y = numpy.array(y).reshape(-1,1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
lr = linear_model.LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
r = round(lr.score(x_test, y_test),2)

prediction_space = numpy.linspace(min(x), max(x)).reshape(-1,1)
plt.scatter(x, y)
plt.plot(prediction_space, lr.predict(prediction_space), color='red')
plt.ylabel('Blood Sugar')
plt.xlabel('bmi')
plt.show()
print("The MSE/mean squared error (quality of model) is: " + ('{0:.6f}').format(mse))
print("The R-squared score (goodness of fit) is: " + str(r))
print("The MSE is really high and the R-squared score is pretty low so the regression isn't really good at representation of the data")
```

```python
# Exercise 5 Part 2
print("Linear Regression Performance")
print("The MSE/mean squared error (quality of model) is: " + ('{0:.6f}').format(mse))
print("The R-squared score (goodness of fit) is: " + str(r))

data = pandas.read_csv('Dataset heart-disease-dataset.csv')
data.replace('?', numpy.nan, inplace=True)
data = data.astype(float)
data.fillna(data.mean(), inplace=True)

x = data.drop('result', axis=1)
y = data['result']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

lor = linear_model.LogisticRegression()
lor.fit(x_train, y_train)

y_pred2 = lor.predict(x_test)
mse_lor = mean_squared_error(y_test, y_pred2)
r_lor = round(lor.score(x_test, y_test),2)

print("\nLogistic Regression Performance")
print("The MSE/mean squared error (quality of model) is: " + ('{0:.6f}').format(mse_lor))
print("The R-squared score (goodness of fit) is: " + str(r_lor))
print("\nUsing a Logistic Regression model is more effective than the Linear Regression model as shown from the lower MSE and higher R-squared score")
```
