```python
#Question 1
import pandas
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy

data = pandas.read_csv('heart-disease-dataset.csv')
data.replace('?', numpy.nan)
data.dropna()

df = data[['age', 'chol']]
X = df.iloc[:,:].values

kmeans = KMeans(n_clusters = 5, random_state = 42)
y_means = kmeans.fit_predict(X)

plt.scatter(X[y_means == 0,0], X[y_means == 0,1], s = 100, c='red', label='0')
plt.scatter(X[y_means == 1,0], X[y_means == 1,1], s = 100, c='orange', label='1')
plt.scatter(X[y_means == 2,0], X[y_means == 2,1], s = 100, c='yellow', label='2')
plt.scatter(X[y_means == 3,0], X[y_means == 3,1], s = 100, c='green', label='3')
plt.scatter(X[y_means == 4,0], X[y_means == 4,1], s = 100, c='blue', label='4')
plt.title("K-means Clustering")
plt.xlabel('Age')
plt.ylabel('Chol')
plt.legend()
plt.show()
```

```python
#Question 2
#!pip install --upgrade kneed
from kneed import KneeLocator

sse = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, random_state=42)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

plt.plot(range(1,11), sse)
plt.title('Number of Clusters vs. SSE')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.show()

k1 = KneeLocator(range(1,11), sse, curve='convex', direction='decreasing')
print("The elbow value (best k value) is", k1.knee)
```

```python
#Question 3
kmeans = KMeans(n_clusters = 3, random_state = 42)
y_means = kmeans.fit_predict(X)

plt.scatter(X[y_means == 0,0], X[y_means == 0,1], s = 100, c='red', label='Group A')
plt.scatter(X[y_means == 1,0], X[y_means == 1,1], s = 100, c='blue', label='Group B')
plt.scatter(X[y_means == 2,0], X[y_means == 2,1], s = 100, c='purple', label='Group C')
plt.title("K-means Clustering")
plt.xlabel('Age')
plt.ylabel('Chol')
plt.legend()
plt.show()
```

```python
#Question 4
import scipy.cluster.hierarchy as sch

dendrogrm = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title("Dendrogram")
plt.xlabel("X")
plt.ylabel("Euclidean distance")
plt.show()
```

```python
#Question 5
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters = 3, affinity = "euclidean", linkage = "ward")
y_hc = hc.fit_predict(X)

plt.scatter(X[y_hc == 0,0], X[y_hc == 0,1], s = 100, c='red', label='Cluster 0')
plt.scatter(X[y_hc == 1,0], X[y_hc == 1,1], s = 100, c='blue', label='Cluster 1')
plt.scatter(X[y_hc == 2,0], X[y_hc == 2,1], s = 100, c='purple', label='Cluster 2')
plt.title("Agglomerative Clustering")
plt.xlabel('Age')
plt.ylabel('Chol')
plt.legend()
plt.show()
```

```python
#Question 6
from sklearn.metrics import silhouette_score
kms = silhouette_score(X, labels=kmeans.labels_, metric='euclidean')
ags = silhouette_score(X, labels=hc.labels_, metric='euclidean')

print("The silhouette score for the k-means clustering is " + str(kms))
print("The silhouette score for the agglomerative clustering is " + str(ags))

print("\nThe silohette score of k-means clustering is higher than the agglomerative clustering. The silouette score represents how accurate the clustering is, so in this case, the k-means clustering is a more accurate representation of the dataset") 
```
